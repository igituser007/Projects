# **Importing libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
```

# *Loading Dataframe and displaying it*

```python
df= pd.read_csv('data-export (1).csv')
```

# *EDA*

```python
df.columns = df.iloc[0]
df.head()                                
```

```python
df.columns = df.iloc[0]
df=df.drop(index=0).reset_index(drop = True)
df.columns = ["channel group" , "DateHour" , "Users" , "Sessions" , "Engaged sessions", "Average engagement time per session" ,	"Engaged sessions per user" , "Events per session" , "Engagement rate" ,	"Event count"]
#removing the duplicated row
```

```python
df.head()
```

```python
df.dropna #dropping null values
```

```python
df.info()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H" ,errors='coerce')
df.head()
```

```python
df.info()
```

```python
numeric_cols = df.columns.drop(["channel group" , "DateHour"])            #leaving these 2 columns all other columns must be  numeric data 
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors='coerce')     
df["Hour"] = df["DateHour"].dt.hour
```

# ***Visualizations***

# Sessions and users overtime

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and users overtime")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.show()
```

# Total users by channels

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x="channel group", y="Users", estimator=np.sum ,palette="viridis")
plt.title("Total users by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Average engagement time by channel

```python
plt.figure(figsize= (10,5))
sns.barplot(data=df,x="channel group", y="Average engagement time per session", estimator=np.mean ,palette="magma")  # here mean is used as estimator cuz we want average
plt.title("Total users by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Organic Video has the highest average engagement time per session, but also a very large error bar, indicating that the data has high variability.

# Referral has a smaller error bar, indicating more consistent engagement times among users from that channel.

# If the team works more on email category it will engage more users

# Engagement Rate Distribution by channel

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df , x="channel group" , y="Engagement rate", palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

Each box represents the interquartile range (IQR) – the middle 50% of values.

The line inside each box is the median.

Dots outside the whiskers are outliers.

# Engaged vs non-engaged sessions

# total sessions - engaged sessions =  non-engaged sessions

```python
session_df = df.groupby("channel group")[["Sessions", "Engaged sessions"]].sum().reset_index()
```

```python
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]
session_df_melted =session_df.melt(id_vars="channel group", value_vars=["Engaged sessions" , "Non-Engaged"])
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=session_df_melted,x="channel group", y="value" , hue="variable")
plt.title("Engaged vs non- engaged sessions")
plt.xticks(rotation=45)
plt.show()
```

# traffic by hour and channel

```python
heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5 , annot=True , fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("channel group")
plt.ylabel("Hour")
plt.show()
```
